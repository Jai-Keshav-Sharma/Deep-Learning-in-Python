```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import *
from keras.models import Model
```

```python
model = Model(inputs=[x, y], outputs=hidden_7)
```

```python
x = Input(shape=(32,))
```

```python
y = Input(shape=(128,))
```

```python
hidden_1 = Dense(8, activation='relu')(x)
hidden_2 = Dense(4, activation='relu')(hidden_1)
hidden_3 = Dense(64, activation='relu')(y)
hidden_4 = Dense(32, activation='relu')(hidden_3)
hidden_5 = Dense(4, activation='relu')(hidden_4)
```

```python
combine = concatenate([hidden_2, hidden_5])
```

```python
hidden_6 = Dense(2, activation='relu')(combine)
hidden_7 = Dense(1, activation='linear')(hidden_6)
```
